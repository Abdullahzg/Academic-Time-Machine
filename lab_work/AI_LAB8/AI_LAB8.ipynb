```python
def flatten_list(primary_list, final_list):
  for item in primary_list:
    if type(item)==list:
      # by using a recursive call on any list inside the list, this function will flatten any level of nested lists
      final_list.append(flatten_list(item, final_list))
    else:
        final_list.append(item)
  return final_list


def flatten_list_wrapper(given_list):
  final_list=list()
  final_list=flatten_list(given_list, final_list)
  # removing leftover empty lists
  for item in final_list:
    if type(item)==list:
      final_list.remove(item)

  return final_list
   
# test_list=[1,[2],[[3],[4],5],6]

# flatten_list_wrapper(test_list)
```
