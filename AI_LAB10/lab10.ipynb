```python
"""Logic Gate based Perceptron"""
import numpy as np
  
# AND Logic Function
def AND_Function(x):
    for i in x:
      if i<=0:
        return 0
    return 1
  
def OR_Function(x):
    for i in x:
      if i > 0:
        return 1
    return 0
# testing the Perceptron Model
test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])
  
print("Results of AND Function")  
print(AND_Function(test1))
print(AND_Function(test2))

print("Results of OR Function")  
print(OR_Function(test3))
print(OR_Function(test4))
```
