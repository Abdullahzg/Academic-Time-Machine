```python
def isPalindrome(givenStr):
  givenStr=givenStr.casefold()
  newStr=''
  for i in givenStr:
    if i>='a' and i<='z':
      newStr += i
  #print(newStr)
  strlen=len(newStr)
  for i in range(strlen//2):
    if newStr[i]!=newStr[strlen-i-1]:
      return False
  return True

givenStr='A man, a plan, a canal: Panama'
if isPalindrome(givenStr) is True:
  print('true')
else:
  print('false')
givenStr='race a car'
if isPalindrome(givenStr) is True:
  print('true')
else:
  print('false')
```
